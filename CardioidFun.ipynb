```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.axes(projection='polar')

theta=np.arange(0, (2*np.pi), 0.01)
r=2*np.sin(10*theta) #10 petalflower

plt.polar(theta,r,'-')

plt.title("Flowers",fontsize=16)
plt.show()
```

```python
plt.axes(projection='polar')

theta=np.arange(0, (2*np.pi), 0.01)

b=1
for a in [-2,-1,-.5,0,.5,1,3]: #loop, cardioid, dimpled, convex
    r=a+b*np.sin(2*theta)
    plt.polar(theta,r,'-')
    plt.show()
```
