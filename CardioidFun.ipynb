# Python doesn't handle negative radii the "right" way 

(To allow starting the radial axis at any value: https://stackoverflow.com/questions/42982290/polar-plot-of-a-function-with-negative-radii-using-matplotlib)

# All angles must be between 0 and 2pi

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.axes(projection='polar')

theta=np.arange(-np.pi/2, np.pi/2, 0.01)
r=theta #raindrop

plt.polar(theta,r,'-')
plt.ylim(r.min(),r.max())
plt.yticks([r.min(),0,r.max()])
plt.xticks([0,np.pi/2,3*np.pi/2])
plt.title("Raindrop?",fontsize=16)
plt.show()
```

# So we have to do this:

```python
plt.axes(projection='polar')

theta=np.arange(-np.pi/2, np.pi/2, 0.01)
r=theta #raindrop

plt.polar(theta+(r<0)*np.pi,abs(r),'-')
plt.ylim(0,abs(r).max())
plt.yticks([0,abs(r).max()/2,abs(r).max()])
plt.xticks([0,np.pi/2,3*np.pi/2])
plt.title("Raindrop!",fontsize=16)
plt.show()
```

```python
theta=np.arange(0, (2*np.pi), 0.01)
r=2*np.sin(10*theta)
plt.polar(theta+(r<0)*np.pi,abs(r),'-')
plt.title("Flowers",fontsize=16)
plt.show()
```

```python
theta=np.arange(0, (2*np.pi), 0.01)
b=1
for a in [.5,1,1.5,3]: #loop, cardioid, dimpled, convex
    r=a+b*np.sin(theta)
    plt.polar(theta+(r<0)*np.pi,abs(r),'-')
    if a==1:
        plt.title("Cardioid",fontsize=16)
    plt.show()
```

```python
theta=np.arange(0, (2*np.pi), 0.01)
b=1
for a in [.5,1,1.5,3]: #loop, cardioid, dimpled, convex
    r=a+b*np.sin(2*theta)
    plt.polar(theta+(r<0)*np.pi,abs(r),'-')
    plt.show()
```
