```python
#pip3.7 install opencv-python
import os
import cv2
import glob
import matplotlib.pyplot as plt
```

```python
#images are in current folder in a folder called frames:

#print(os.getcwd())
for filepath in glob.glob('frames/*.jpg'):
    #gets files in random order
    print(filepath)
#can also search subdirectories:
#glob.glob('**/*.jpg', recursive=True)
```

```python
files=os.listdir('frames') 
for filename in sorted(files, key=str):
        filepath = os.path.join('frames', filename)
        print(filepath)
        img=cv2.imread(filepath)
        
        #display color image
        b,g,r=cv2.split(img)
        #cv2 uses BGR order, need to convert to RGB order
        plt.imshow(cv2.merge([r,g,b]))
        plt.show()
        
        #convert color to greyscale
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        #add vmin vmax if you don't want matplotlib to change pixel scale intensity
        plt.imshow(gray,cmap='gray', vmin=0, vmax=255)
        plt.show()
```
