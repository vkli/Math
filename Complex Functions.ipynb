# 1. Domain Coloring Plots

```python
#Adapted from 
#https://www.mi.fu-berlin.de/en/math/groups/ag-geom/teaching/14_SoSe/VisualisierungABV/downloads/ABV_Visualization_2014-Lab-Domain_Coloring.pdf
#and
#https://commons.wikimedia.org/wiki/User:Jan_Homann/Mathematics
#and
#https://github.com/hernanat/dcolor/blob/master/dcolor.py

# Used to embed plots into notebook view
# Note: Remove this line for IPython version < 1.0
%matplotlib inline
# We need a plot function from matplotlib
import matplotlib.pyplot as plt
# We may need a color conversion function
from matplotlib.colors import hsv_to_rgb
# Use numpy for arrays, much faster than python arrays/lists
import numpy as np
# Standard complex functions and values
# Doesn't work with meshgrid, use numpy instead
from cmath import pi, sin, cos, exp, log

def normalize(arr):
    """Used for normalizing data in array based on min/max values"""
    arrMin = np.min(arr)
    arrMax = np.max(arr)
    arr = arr - arrMin
    return arr / (arrMax - arrMin)

def my_color_scheme(z):
    # get polar coordinates, angle in [0,2pi]
    theta = np.angle(z)
    r = np.abs(z)
    # Conversion from hsv (hue, saturation, lightness) to rgb color space
    # This is needed since plotting function takes rgb pixel colors.
    theta=normalize(theta)
    r=normalize(r)
    return hsv_to_rgb(np.dstack((theta,1-r,0*r+1)))
```

# f(z)=z

```python
#n x n points
n=2000
a=np.linspace(-10,10,n)
b=np.linspace(-10,10,n)
x,y=np.meshgrid(a,b)
z=x+y*1J
w=z
img=my_color_scheme(w)
plt.imshow(img,extent=[-1,1,-1,1])
plt.xlabel('Re',fontsize=12)
plt.xticks([-1,0,1],('-10','0','10'),fontsize=12)
plt.ylabel('Im',fontsize=12)
plt.yticks([-1,0,1],('-10','0','10'),fontsize=12)
plt.title(r'$z$',fontsize=16)
plt.show()
```

# f(z)=Log(z)

```python
'''principal branch of complex logarithm'''

#n x n points
n=2000
a=np.linspace(-10,10,n)
b=np.linspace(-10,10,n)
x,y=np.meshgrid(a,b)
z=x+y*1J
r=np.abs(z)
theta=np.angle(z)
w=np.log(r)+theta*1J
img=my_color_scheme(w)
plt.imshow(img,extent=[-1,1,-1,1])
plt.xlabel('Re',fontsize=12)
plt.xticks([-1,0,1],('-10','0','10'),fontsize=12)
plt.ylabel('Im',fontsize=12)
plt.yticks([-1,0,1],('-10','0','10'),fontsize=12)
plt.title(r'$z$',fontsize=16)
plt.show()
```
